```python
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'x': [i/10 for i in range(1,101)], 
                  'Squared':[(i/10)**2 for i in range(1,101)],
                     'Root':[np.sqrt(i/10) for i in range(1,101)]})
```

```python
fig = px.line(data_frame=df, x='x', y=['Squared', 'Root'])
# fig = px.line(x=[i/10 for i in range(1,101)], y=)
fig.update_layout(title='Squared', title_x=0.5)
fig.show()
```
